```python
#Definindo instanciando classe

class Carro:

    def __init__(self, marca, modelo, cor):
        self.marca = marca
        self.modelo = modelo
        self.cor = cor
        self.velocidade = 0

    def acelerar(self):

        self.velocidade += 10

        print(f"Velocidade atual {self.velocidade} km/h")

    def frenar(self):
        self.velocidade -= 10

        if self.velocidade < 10:

            self.velocidade = 0

        print(f"Velocidade atual {self.velocidade} km/h")

    def exibir_info(self):
      print(f"Marca: {self.marca}, modelo: {self.modelo}, cor: {self.cor}, velocidade: {self.velocidade} km/h")



lista_carros = []

while True:
    print("\n---Menu---:")
    print("1 - Adicionar carro")
    print("2 - Exibir informaçoes do carro")
    print("3 - Acelerar um carro")
    print("4 - Frenar um carro")
    print("5 -Sair")


    escolha = input("Escolha uma Opção: ")

    if escolha == "1":
        print("\n---Adicionar carro---:")
        marca = input("Insira a marca: ")
        modelo = input("Insira o modelo: ")
        cor = input("Insira a cor: ")

        novo_carro = Carro(marca, modelo, cor)

        lista_carros.append(novo_carro)

    elif escolha == "2":
        print("\n---Exibir informaçoes do carro---:")

        if lista_carros:
            for carro in lista_carros:
                carro.exibir_info()
        else:
            print("Nenhum carro encontrado")

    elif escolha == "3":
        print("\n---Acelerar um carro---:")

        modelo = input("Insira o modelo do carro que deseja acelerar: ")

        for carro in lista_carros:
            if carro.modelo == modelo:
                carro.acelerar()
                break
            else:
                print("Modelo do carro não encontrado")

    elif escolha == "4":
        print("\n---Frenar um carro---:")

        modelo = input("Insira o modelo do carro que deseja frenar: ")

        for carro in lista_carros:

            if carro.modelo == modelo:
                    carro.frenar()
                    break
            else:
                print("Modelo do carro não encontrado")

    elif escolha == "5":
        print("\n---Saindo---:")
        break
    else:
        print("Opção inválida")





```
