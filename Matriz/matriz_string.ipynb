###Exercicio urna eletrônica

```python
#Crie um programa que simule uma urna eletronica simples usando matrizes

urna = [["João", 0], ["Pedro", 0], ["Alice", 0]]

for i in range(4):

    voto = input("Digite o nome do candidato que deseja votar (João, Pedro, Alice):")

    encontrado = False

    for candidato in urna:
        if candidato[0] == voto:

            candidato[1] += 1

            encontrado = True
            break

if not encontrado:
    print("Voto nulo!")


print("\nResultados:")

votos_maximo = -1
vencedor = ""

for candidato in urna:
    print(f"{candidato[0]}:{candidato[1]} votos")

    if candidato[1] > votos_maximo:
        votos_maximo = candidato[1]
        vencedor =  candidato[0]

print(f"O vencedor é {vencedor} com {votos_maximo} votos")

print("")

```

##RESERVA DE ASSENTO NO CINEMA

```python
#SOLUÇÃO

#Inicia uma lista vazia para a matri cinema
cinema = []

#Loop externo para percorrer as fileiras

for i in range(5):

    #Iniciar uma lista vazia para a fileira atual
    fileira = []
    #Loop interno para perrcorrer os assentos dentro da fileira
    for j in range(10):
        #Adicionar um assento disponivel ("D) à fileira atual
        fileira.append("D")
        #Adicionar a fileira completa à matriz do cinema
    cinema.append(fileira)

while True:

#Imprime o menu de opções
    print("\nMenu:")
    print("1. Ver disposição dos assentos")
    print("2. Reservar assento")
    print("3. Sair")

#Solicita ao usuario que escolha uma opção
    escolha = input("Escolha uma opção: ")

#Se o usuario escolher a opção 1
    if escolha == "1":

#Inicia um loop para percorrer cada fileira da matri cinema
        for fileira in cinema:
            #Inicia um interno para percorrer da assento da fileira atual
            for assento in fileira:
                #Imprime o statut do assento (D para disponivel e R para reservado), sem quebrar linha
                print(assento, end=" ")
            #imprime uma quebra de linha após exibir todos os assentos de uma fileira completo
            print()
     #Se o usuario escolher a opção 2
    elif escolha == "2":
            fileira = int(input("Digite o numero da fileira / linha (0-4): "))
            assento = int(input("Digite o numero da assento / coluna (0-9): "))

            if cinema[fileira][assento] == "D":
                cinema[fileira][assento] = "R"
                print("Assento reservado com sucesso!")
            else:
                print("Assento já está reservado ou ocupado!!")


    elif escolha == "3":
        print("Obrigado por usar nosso sistema!")
        break
    else:
        print("Opção invalida!")


print()
```
