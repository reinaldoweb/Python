```python
"""
    Listas e Strings
    Conversão de Strings para Listas: list(), split()
    Conversão de Lista para string: join()
"""

# Exemplo prático
# Conversão de Strings para Listas

s = 'O rato roeu a roupa do rei de roma'
list_s = list(s)

print(list_s)

#B) split()

s = 'O rato roeu a roupa do rei de roma'
list_s = s.split()

print(list_s)

data = "12/10/2024"
list_data = data.split('/')

print(list_data)
print()
#Conversão de listas para strings

#Usando join()

s = ['O', 'rato', 'roeu', 'a', 'roupa', 'do', 'rei', 'de', 'roma']

palavras_juntadas = ' '.join(s)

print(palavras_juntadas)

```

```python
# Exercício Lista de Compras
lista_compras = []

while True:
    print("Selecione a opção desejada: \n")
    print("1. Adicionar item a lista")
    print("2. Remover item da lista")
    print("3. Exibir item da lista")
    print("4. Sair")

    opcao = int(input("Escolha uma opção: "))

    if opcao == 1:
        item = input("Adicione um item a lista: ")
        lista_compras.append(item)
        print("O item adicionado a lista:", item)
    elif opcao == 2:
        if len(lista_compras) == 0:
         print("Lista vazia. Nada para remover")
        else:
            item = input("Digite o item que deseja rmover da lista: ")
            if item in lista_compras:
                lista_compras.remove(item)
                print("Item removido da lista")
            else:
                print("O item não existe na lista")

    elif opcao == 3:
        print("Lista de compras")

        for item in lista_compras:
            print("-", item)

    elif opcao == 4:
        print("Saindo...")
        break
    else:
        print("Opção inválida. Tente novamente.")
```

```python
# Exercicio classificação de numeros

"""
Inserir numeros inteiros um a um
armazenar em uma lista
Classificar em ordem crescente
Imprimir lista classificada
"""

#Inicializa a variavel vazia

numeros = []

# Loop for para repetir o bloco 5 vezes

for i in range(5):
    #Pedi ao usuário para inserir um numero
    numero_digitado = int(input("Digite um nomero: "))
    #Adiciona o numero na lista com append()
    numeros.append(numero_digitado)
#Classifica o numero na lista em ordem crescente
numeros.sort()
#Imprime a lista classificada
print("Números ordenado: ",numeros)
```
